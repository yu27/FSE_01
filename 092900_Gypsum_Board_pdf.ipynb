```python
raw = parser.from_file('092900-Gypsum Board.pdf')
text = raw['content']
print(text)
```

```python
#pattern = re.compile("SECTION[ ]?[0-9].*[ ]?-.*")
p_section = re.compile("SECTION \d{6} - [A-Z]*")
sections = re.findall(p_section, text)
if (len(sections) > 0):
    section = sections[-1]
    print(section)
else:
    print("Section cannot be found")

section_num = section.split()[1]
section_name = section.split()[-1]
print("Section Number: ", section_num)
print("Section Name: ", section_name)
```

```python
pattern = re.compile("PART [2-3] - [a-zA-Z]*")
str2 = re.findall(pattern, text)
print(str2[0])
```

```python
text = text[text.find(str2[0]):] #index of str2[0] to the end of the text
print(text)
```

```python
p_category = re.compile("2.[3-4] [A-Z -]*[\n]")
categories = re.findall(p_category, text)
print (categories)
```

```python
index = 0
for category in categories:
  if "INTERIOR GYPSUM BOARD" in category:
    print("Process all materials under INTERIOR GYPSUM BOARD")
    index = categories.index(category)
    str_ws = text[text.find(categories[index]):text.find(categories[index+1])]
    #print(str_ws)
    p_material = re.compile(r"[^D,0-9][.][a-zA-Z, -]*[:]")
    materials = re.findall(p_material, str_ws)
    print("Materials:",materials) #returns list of lettered subheadings
    for index_m in range(len(materials)-1):
      m_name = ''
      m_type = ''
      m_location=''
      m_exposure=''
      m_level=''

      m_name = materials[index_m].split(".")[-1][:-1].strip()
      str_material = str_ws[str_ws.find(materials[0]):]
      m_types = re.findall('Type X', str_material)
      #print(m_types) #list of types
      for m_type in set(m_types):
          m = re.search('Exterior|Interior', str_material)
          if m:
              m_location = m.group(0)
          m = re.search('Exposure [1-3]', str_material)
          if m:
              m_exposure = m.group(0)
          '''m = re.search('Level [1-9]', str_material)
          if m:
              m_level = m.group(0)'''
      #m = re.search('Exterior|Interior', str_material)
      #if m:
         # m_location = m.group(0)
          print("Material Name: ", m_name)
          print("Material Type: ", m_type)
          #print("Material Location: ", m_location)
          #print("Material Exposure: ", m_exposure)
          print("\n")
```

```python
p_category = re.compile("3.[7-8] [A-Z -]*[\n]")
categories = re.findall(p_category, text)
print (categories)
```

```python
index = 0
for category in categories:
    if "FINISHING GYPSUM BOARD" in category:
        #print("Process all materials under FINISHING GYPSUM BOARD")
        index = categories.index(category)
        str_ws = text[text.find(categories[index]):text.find(categories[index+1])]
        #print(str_ws) #section 3.7 of text
        p_material = re.compile(r"[D-E][.][a-zA-Z, -]*[:]")
        materials = re.findall(p_material, str_ws) #list of lettered subheadings
        l_material = re.compile(r"Level [1-5]")
        levels = re.findall(l_material, str_ws)
        #print(materials)
        #print(levels)
        #for index_m in range(len(materials)-1):
        for level in levels:
          m_type = "Gypsum Board Finish Levels"  
          m_name = "FINISHING GYPSUM BOARD"
          m_location =''
          m_exposure =''
          m_level = level
          
          str_material = str_ws[str_ws.find(materials[0]):str_ws.find(materials[1])]
          location = re.search('Insert locations', str_material)
          if location:
            m_location = location.group(0)
          print("Material Name: ", m_name)
          print("Material Type: ", m_type)
          print("Material Location: ", m_location)
          print("Material Exposure: ", m_exposure)
          print("Material Level: ", m_level)
          print("\n")
```
